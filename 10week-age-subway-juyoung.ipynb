```python
import matplotlib.pyplot as plt

x = [0, 2, 4, 5]
y = [5, 10, 15, 20]
plt.bar(x, y)
plt.savefig('막대그래프.png')
plt.savefig('막대그래프_dpi_50.png', dpi=50)
plt.savefig('막대그래프_dpi_200.png', dpi=200)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0, 2, 4, 5]
y = [5, 10, 15, 20]
plt.bar(x, y)
plt.savefig('막대그래프_facecolor.png', facecolor='skyblue')
plt.show()
```
