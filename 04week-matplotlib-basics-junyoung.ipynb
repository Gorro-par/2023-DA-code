## 색상 및 각종 속성(스타일) 지정

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40],'skyblue',label='skyblue')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10],'pink',label='pink')

plt.title('legend')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40],'skyblue',linestyle='--',marker='d',label='diamond')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10],'pink',ls=':',marker='^',label='tyriangle up')

plt.title('Marker')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import font_manager

plt.plot([-2, -1, 0, 1,2,3], [20, 30, 15, 40, 45, 35], label='선 1')
plt.plot([-2, -1, 0, 1,2,3], [25, 10, 35, 20, 15, 25], label='선 2')
plt.rc('font', family='Malgun Gothic')
plt.title('선 그래프 예제')
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x, y)

plt.grid(True)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title('삼각함수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o')
plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o-')
plt.plot(x2, y + 0.5, 'o:')
plt.ylim([-0.5, 1])
plt.show()
```
